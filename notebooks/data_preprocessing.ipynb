## Data Pre-processing Pipeline

Based on the analysis, the following actions are conducted in the data cleaning pipeline:

- Drop the `customerID` column, as it is not needed for modeling.
- Convert `TotalCharges` to numeric (float) type, coercing errors to NaN.
- Fill missing values in `TotalCharges` by calculating `tenure` multiplied by `MonthlyCharges`.
- Convert the `Churn` column from categorical labels (`'Yes'`/`'No'`) to binary integers (`1`/`0`).
- One-hot encode categorical features to convert them into numerical format suitable for modeling, dropping the first category to avoid multicollinearity.
- Convert boolean columns resulting from encoding to integer type (0 or 1).
- Save the cleaned data to a CSV file at the specified output path.

These steps ensure the dataset is clean, consistent, and ready for exploratory data analysis and model training.

```python
def cleaning_pipeline(data: pd.DataFrame, output_path: str) -> pd.DataFrame:
    """Data cleaning and processing pipeline

    Args:
        data (pd.DataFrame): input dataset that needs to be pre-processed
        output_path (str): location of where to save the csv file
    Returns:
        data (pd.DataFrame): cleaned dataset
    """

    #Drop customerID
    data.drop(columns=['customerID'], inplace=True)

    #Convert TotalCharges type
    data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
    #Fill missing value of TotalCharges with tenure x MonthlyCharges
    data['TotalCharges'] = data['TotalCharges'].fillna(data['tenure'] * data['MonthlyCharges'])

    #Convert Churn to binary
    data['Churn'] = data['Churn'].map({'No':0, 'Yes':1}) 

     #Convert other strictly binary yes/no columns
    binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
    for col in binary_cols:
        data[col] = data[col].map({'Yes': 1, 'No': 0})

    internet_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection','TechSupport', 'StreamingTV', 'StreamingMovies']
    phone_cols = ['MultipleLines']

    # Replace "No internet service" with "No"
    data[internet_cols] = data[internet_cols].replace('No internet service', 'No')
    # Replace "No phone service" with "No"
    data[phone_cols] = data[phone_cols].replace('No phone service', 'No')

    
    data.to_csv(output_path, index=False)
    return data
```

```python
output_path = '../data/cleaned_data.csv'
cleaning_pipeline(data, output_path)
```
